# openai

```python
from openai_scripts import openai_embed_local
embeddings = openai_embed_local()
```

# AStraDB

```python
from astradb_scripts import astradb_start
embedding_model = 'BAAI/bge-large-en-v1.5'
collection_name = embedding_model.replace('/', '_').replace('-', '_').replace('.', '_')
vstore = astradb_start(
    **{
        'embeddings': embeddings,
        'collection_name': collection_name
    }
)
```

## PDF add to astradb

```python
from astradb_scripts import astradb_add_pdf
astradb_add_pdf(
    **{
        'vstore': vstore,
    }
)
```
